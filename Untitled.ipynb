```python
df.isnull().sum()
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
```

```python
missing_value_df=missing_value_df[missing_value_df != 0].sort_values('percent_missing', inplace=False)
```

```python
missing_value_df[missing_value_df != 0].plot(kind='bar')
```
